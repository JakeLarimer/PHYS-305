### 1. Finite Differences

Given function $f(x)$, want to evaluate 

$$
\frac{df}{dx}\Big|\normalsize_{x_0} \approx \frac{f(x_0+h)-f(x_0)}{h}
$$

when $h$ is small.

Two sources of error
- a. Truncation error
- b. Round off error

##### a. Truncation error:

Do a Taylor Expansion

$$
f(x_0+h)\approx f(x_0) + hf'(x) + \frac{1}{2}h^2f''(x)+\frac{1}{6}h^3f'''(x) + ... \text{(eqn. 2)}
$$

Plugging this expansion back in

$$
\tilde{f}'(x) \approx f' + \frac{1}{2}hf'' + ...
$$

where $\tilde{f}'$ is numerical, $f'$ is true. Difference is truncation error.

$$
e_t(h)\approx hf'' \quad\text{(eqn. 3)}
$$

##### b. Round off error

This is a machine's limitation to represent floating point numbers, $e_m$.

Consider two points, $x$ and $x+h$. 

$$
[(x+h) - x] \not\simeq h = e_m\alpha
$$

$$
e_r \approx \frac{e_mf(x)}{h} \quad\text{(eqn. 4)}
$$

where $e_r$ is Round off error.

$$
\text{Net} = e_r + e_t = hf''+\frac{e_mf}{h}
$$

We want to find a value of $h$ such that the net error is minimized by differentiating with respect to $h$ and setting to zero.

$$
\frac{d(\text{Net})}{dh}=0
$$

After solving : 

$$
h_{min} \sim \sqrt{e_m\left(\frac{f}{f''}\right)} \quad\text{(eqn. 5)}
$$

Assume :
$$
\sqrt{\frac{f}{f''}} = x \quad (\text{curvature scale})
$$

Find total error for $h_{min}$
$$
= e_t(h_{min}) + e_r(h_{min})
$$

Fractional error
$$
= \frac{e_r(h_{min}) + e_t(h_{min})}{f'}
$$

$$
= \frac{\frac{e_mf}{\sqrt{e_m\frac{f}{f''}}} + \sqrt{e_m\frac{f}{f''}}f''}{f'} = \frac{e_mf + e_m\frac{f}{f'}}{f'\sqrt{e_m\frac{f}{f''}}}
$$

$$
= \frac{\sqrt{e_mff''}}{f'}
$$

$f$, $f'$, $f''$ are of similar scale

$$
= \sqrt{e_m}
$$

##### Two - Sided Derivative

$$
f'(x) \sim \frac{f(x+h) - f(x-h)}{2h}
$$

Recall Taylor Series:

$$
f(x+h) = f(x) + hf'(x) + \frac{1}{2}h^2f''(x)
$$

$$
f(x-h) = f(x) - hf'(x) + \frac{1}{2}h^2f''(x)
$$

Consider

$$
f(x+h)-f(x) \quad\text{vs.}\quad f(x+h)-f(x-h)
$$

Works out to 

$$
e_t \sim h^2f'''
$$

$$
h_{min} \sim \left(\frac{e_mf}{f'''}\right)^{1/3}
$$

$$
\text{fractional error} = e_m^{2/3}
$$
